# Customer Churning Predictions Using TensorFlow

Import Data

```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
# csv file dataframe
df = pd.read_csv('Churn.csv')
```

```python
# Clean data and set x and y variables
X = pd.get_dummies(df.drop(['Churn', 'Customer ID'], axis=1))
y = df['Churn'].apply(lambda x: 1 if x=='Yes' else 0)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2)
```

```python
X_train.head()
```

```python
y_train.head()
```

Import TensorFlow API

```python
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Dense, Input # This allows us to add more layers of neurons
from sklearn.metrics import accuracy_score
```

Build and Compile Model

```python
input_shape = (len(X_train.columns),)

model = Sequential()
model.add(Input(shape=input_shape))
model.add(Dense(units=32, activation='relu')) # 32 neurons in this layer, hockeystick function
model.add(Dense(units=64, activation='relu')) # 64 neurons in this layer
model.add(Dense(units=1, activation='sigmoid')) # 1 neuron in this layer
```

```python
model.compile(loss='binary_crossentropy', optimizer='sgd', metrics=['accuracy'])
```

Fit, Predict, and Evaluate with 500 Epochs

```python
model.fit(X_train, y_train, epochs=500, batch_size=32)
```

```python
y_hat = model.predict(X_test)
y_hat = [0 if val < 0.5 else 1 for val in y_hat]
```

```python
accuracy_score(y_test, y_hat)
```
